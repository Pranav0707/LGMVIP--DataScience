```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/butterfly - Copy.jpg')
cv2_imshow(img)
```

```python
gray_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
intensity_reduced=255-gray_image
cv2_imshow(intensity_reduced)
```

```python
blurred_image=cv2.GaussianBlur(intensity_reduced,(21,21),0)
blurred_intensity_reduced=255-blurred_image
pencil=cv2.divide(intensity_reduced,blurred_image,scale=256.0)
cv2_imshow(pencil)
```

```python
cv2_imshow(img)
cv2_imshow(pencil)
```
